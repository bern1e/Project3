```python
#Import packages
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Importing individual Excel Files
stem_indiv_outcomes = pd.read_excel("STEM Outcomes Individual Data 2002 - 2018 (anon).xlsx")
stem_outcomes = pd.read_excel("STEM Outcomes Grades 2002 - 2018 (anon).xlsx")
codes = pd.read_excel("Majors Titles and CIP Codes.xlsx")
```

```python
#Merging Data Sets Together
stem_outcomes_merged = stem_outcomes.merge(stem_indiv_outcomes, on = "Student ID", how = "inner")
stem_outcomes_merged
```

```python
#Merging Subject and Course Number for Later Analysis of Course Distribution within Major. COnfirm that this needs to be done
stem_outcomes_merged["Course"] = stem_outcomes_merged["Subject"] + stem_outcomes_merged["Course Number"].apply(str)
stem_outcomes_merged
```

Pick a popular course from the bar graph you built in Exercise 2 and generate a normalized histogram of the grades in that class, but only for students that eventually earned the degree you picked in Exercise 2. Compare to a grade histogram of students that did not earn that degree.

```python
#First let's figure out how many economics majors we have accross the One Major, Double-Major and Triple Major Students:
#Accumulation of Economics Majors
stem_economics = stem_outcomes_merged[(stem_outcomes_merged["Major 1"] == "ECON") | (stem_outcomes_merged["Major 2"] == "ECON") | (stem_outcomes_merged["Major 3"] == "ECON")]
#Number of Economics Majors
stem_politics["Student ID"].nunique()
```

```python
#Examining ECON-102 Outcomes for Economics Majors
ECON102_grades = stem_economics["Grade"][stem_economics["Course"]  == "ECON102"].dropna()
ECON102_grades.plot(kind = 'hist', rwidth = 0.95, bins =16, color = "DarkBlue") #16 Bins for grade levels increasing in .25 increments.
plt.xlabel("Grade Earned")
plt.title("Grade Earned for Economics Majors Who Took ECON102")
```

```python
#Determining Average Grade Earned Across the Courses, this will added to the histogram.
mean_grade_econ102 = ECON102_grades.mean()
mean_grade_econ102
```

```python
mean_grade_econ101 = ECON101_grades.mean()
mean_grade_econ101
```

```python
#Let's Now Normalize Each and See How the dust Settles with the Incorporation of the mean grade as well
plt.hist([ECON102_grades], alpha = 0.5, bins = 16, rwidth = 0.95, density = True, color = "DarkBlue", label = "Economics Major ECON102 Grades")
plt.hist([ECON101_grades], alpha = 0.5, bins = 16, rwidth = 0.95, density = True, color = "Orange", label = "Economics Major ECON101 Grades")
plt.xlabel("Grade Earned")
plt.ylabel("Frequency")
#plt.axvline(average_grade, color='red', linestyle='dashed', linewidth=2, label=f'Average Grade: {average_grade:.2f}') #Consulted GPT here to learn about plotting of vertical line/mean. 
plt.axvline(mean_grade_econ102, color = "DarkBlue", linestyle = "solid", label = "Average Grade of ECON102 Across ECON Majors")
plt.axvline(mean_grade_econ101, color = "Orange", linestyle = "solid", label = "Average Grade of ECON101 Across ECON Majors")
plt.legend()
plt.show()
```

At first glance it appears English majors did quite well in PSYC110. Let's now reincorporate our findings from the Politics majors in addition to students who were neither Enlish or Politics majors. We will normalize each grouping for comparison. 

First, let's create a new variable where both politics majors and english majors are excluded. This will serve as our Non-Major category. 

```python
#
general_psych110 = stem_outcomes_merged["Grade"][(stem_outcomes_merged["Course"] == "PSYC110") &((stem_outcomes_merged["Major 1"] != "POL") &(stem_outcomes_merged["Major 2"] != "POL") &(stem_outcomes_merged["Major 3"] != "POL") &
        (stem_outcomes_merged["Major 1"] != "ENGL") & (stem_outcomes_merged["Major 2"] != "ENGL") & (stem_outcomes_merged["Major 3"] != "ENGL"))].dropna()
general_psych110.plot(kind = 'hist', rwidth = 0.95, bins = 16, color = "Orange")
plt.title("Non-Politics or Economics Degree Students Grade Earned")
plt.xlabel("Grade Earned")
plt.show()
```

```python
#Now Including the Politics Major PSYC110 (Worked with earlier in project) Data we will attempt to normalize and plot the three different groupings. 
plt.hist([psyc110_grades], alpha = 0.3, bins = 16, rwidth = 0.95, density = True, color = "DarkBlue", label = "Politics PSYC110 Grades")
plt.hist([engl_psyc110_grades], alpha = 0.3, bins = 16, rwidth = 0.95, density = True, color = "Maroon", label = "English PSYC110 Grades")
plt.hist([general_psych110], alpha = 0.3, bins = 16, rwidth = 0.95, density = True, color = "Orange", label = "Non-Politics or English Major PSYC110 Grades")
plt.xlabel("Grade Earned")
plt.ylabel("Frequency")
plt.title("Grade Point Earned After Completion of PSYC110")
plt.legend()
plt.show()
```
